## Loading CSV File

> CSV  - Comma Separeted Files   
> CSV have .csv extention

### 1. Using Sep Parameter  
> 1. We can mention our field separator based on the input CSV file structure.  
> 2. In some cases, some different types of special characters are used as separators in CSV, in that situation sep parameter will help us.  

```python
import pandas as pd

data = pd.read_csv('../00_data/nyc_flights.csv', 
                   sep=",")
display(data.head())
```

### 2. Using header Parameter  
> This parameter is used to sepecify a particular row as header.  

```python
import pandas as pd

data = pd.read_csv('../00_data/nyc_flights.csv', 
                   header=1)
display(data.head())
```

### 3. Using names Parameter  
> It is used to rename the columns.  

```python
import pandas as pd

data = pd.read_csv('../00_data/nyc_flights.csv', 
                   names =  ['YEAR',
                             'MONTH',
                             'DAY',
                             'DEP_TIME',
                             'DEP_DELAY',
                             'ARR_TIME',
                             'ARR_DELAY',
                             'CARRIER',
                             'TAILNUM',
                             'FLIGHT',
                             'ORIGIN',
                             'DEST',
                             'AIR_TIME',
                             'DISTANCE',
                             'HOUR',
                             'MINUTE'])
display(data.head())
```

### 4. Using index_col Parameter
> To use a particular columns as index.

```python
import pandas as pd

data = pd.read_csv('../00_data/nyc_flights.csv', 
                   index_col="year")
display(data.head())
```

```python
import pandas as pd

data = pd.read_csv('../00_data/nyc_flights.csv', 
                   index_col= 0)
display(data.head())
```

### 5. Using use_cols Parameter
> Used to load some speciic columns from data.

```python
import pandas as pd

data = pd.read_csv('../00_data/nyc_flights.csv', 
                   usecols=['month', 'day', 'dep_time', 'dep_delay'])
display(data.head())
```
