# Feature Hashing Encoding

### 1. Creating DF

```python
import pandas as pd

# Creating DataFrame with categorical data.
data = pd.DataFrame([['@gmail.com', 0], 
                     ['@nvid.com', 1], 
                     ['@hotmail.com', 0],
                     ['@us-gt.com', 1],
                     ['@yahoo.com', 0],
                     ['@hotmail.com', 0],
                     ['@us-gt.com', 1],
                     ['@yahoo.com', 0]],
                   columns=["email_domine", "spam"])

data
```

```python
from sklearn.feature_extraction import FeatureHasher

n_features = 2
features = [f"fh{n}" for n in range(0, n_features)]
fh = FeatureHasher(n_features=n_features, input_type='string')
r = fh.fit_transform(data['email_domine'])
print(r.toarray())

hasher_df = pd.DataFrame(r.toarray(), columns= features)

data = pd.concat([data, hasher_df], axis=1)
data
```
