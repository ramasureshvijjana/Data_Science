```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.DataFrame([[1, 'red'], [2, 'green'], [3, 'blue'], [4, 'red']], columns=['ID', "color"])
display(data)

label_encoder = LabelEncoder()
data['color'] = label_encoder.fit_transform(data['color'])

display(data)
```

### 1. Data Loading

```python

import pandas as pd

# Loading Data
data = pd.read_csv('E:/DS/DataScience/Dealing_With_categorical_Data/data/Nutritional_Content_of_Cereals/cereal.csv')
display(data.head())
```

```python
print(data['mfr'].unique())
print(data['type'].unique())
```

### 2. Drop null and unwanted values

```python
data.drop(0, inplace=True)
data.dropna(inplace=True)

data.reset_index(drop=True, inplace=True)
```

```python
display(data)
```

### 3. Use LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

coder = LabelEncoder()
data['mfr'] = coder.fit_transform(data['mfr'])
data['type'] = coder.fit_transform(data['type'])

display(data)
```
