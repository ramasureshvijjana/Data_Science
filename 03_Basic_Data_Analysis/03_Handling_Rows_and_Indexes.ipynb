## Handling Rows And Indexes

> Indexes of data is one of the important topics in the data handling.   
> Here will learn about how to see the index numbers and how to set a particular column as index.

```python
import pandas as pd

data = pd.read_csv('E:/DATA_SCIENCE/DS/DataScience/data/nyc_flights.csv')
display(data)
```

### 1. index perameter
> Returns list of index numbers

```python
print(data.index)

print(type(data.index))
```

### 2. set_index() function
> To set a specific column value as index numbers.

```python
data.set_index("day", inplace=True)
display(data)
```

```python
data.set_index(["dep_time", "dep_delay"], inplace=True)
display(data)
```

### 4. drop() function
> Drop some specific rows based on index numbers.

```python
import pandas as pd

data = pd.read_csv('E:/DATA_SCIENCE/DS/DataScience/data/nyc_flights.csv')

data.drop(1, inplace=True)
display(data)
```

```python
data.drop([2,3,4,6], inplace=True)
display(data)
```

```python
print(data.index)
```
