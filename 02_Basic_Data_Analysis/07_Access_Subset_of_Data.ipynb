## Access Subset of Data
> Pandas provides 2 interesting functions iloc and loc for access subset of data.

```python
import pandas as pd

data = pd.read_csv('E:/DATA_SCIENCE/DS/DataScience/data/nyc_flights.csv')
display(data.head())
```

### 1. iloc perameter

```python
first_row_series = data.iloc[0] # Returns 0th row as pandas series.

display(first_row_series)
print(f"Data type is {type(first_row_series)}")
```

```python
first_row_df = data.iloc[[0]] # Returns 0th row as pandas data frame.

display(first_row_df)
print(f"Data type is {type(first_row_df)}")
```

```python
specific_df = data.iloc[[0,1], [2,4]] # Return data frame with mentioned row and column indexes.

display(specific_df)
print(f"Data type is {type(specific_df)}")
```

```python
df_subset_way1 = data.iloc[1:3, 0:3] # 1st row to 2nd row and 0th column to 2nd column.

display(df_subset_way1)
print(f"Data type is {type(df_subset_way1)}")
```

```python
df_subset_way2 = data.iloc[:3, 0:3] # Starting row to 2nd row and 0th column to 2nd column.

display(df_subset_way2)
print(f"Data type is {type(df_subset_way2)}")
```

```python
df_subset_way3 = data.iloc[10:, 0:3] # 10th row to last row and 0th column to 2rd column

display(df_subset_way3)
print(f"Data type is {type(df_subset_way3)}")
```

```python
df_subset_way4 = data.iloc[:, 0:3] # First row to last row and 0th column to 2nd column.

display(df_subset_way4)
print(f"Data type is {type(df_subset_way4)}")
```

### 2. loc perameter

```python
df = pd.DataFrame([[1, 2, 3], [4, 5, 31], [7, 8, 32]],
     index=['cobra', 'viper', 'sidewinder'],
     columns=['max_speed', 'shield', 'weight'])
df
```

```python
df.loc[['cobra','viper']]
```

```python
df.loc['cobra':'viper','max_speed':'shield']
```
