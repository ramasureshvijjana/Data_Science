# Observe Null values

```python
=> Two ways to find NULL values:
    1. Data formate
    2. Graphical formate
    
1. Data Formate:
    1. isnull() or isna()
    2. notnull() or notna() # notnull() is just reverse of isnull()

2. Graphical Formate:
    PACKAGE: import missingno as msno
    
    1. matrix()
    2. bar()
    3. heatmap()
```

## Data Loading.

```python
import pandas as pd

data = pd.read_csv("data/titanic.csv")
```

## 2. Graphical Formate

```python
# Installing missingno package
pip install missingno
```

```python
import missingno as msno

msno.matrix(data)
```

```python
msno.bar(data)
```

```python
msno.heatmap(data)
```
